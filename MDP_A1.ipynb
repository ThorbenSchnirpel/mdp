```python
import matplotlib.pyplot as plt
import math

def y_dot(x, y):
   return (10-(500*y)+(5000*x))

 
def y(x):
   return (10*x+math.exp(-500*x))

 
h = 0.004
x_End = 0.2

steps = x_End/h

x = [steps]
x[0]=0

#explizites Euler-Verfahren
y1 = [steps]
y1[0]=1
y1_diff = [steps]
y1_diff[0] = 0

#RK2-Verfahren
y2=[steps]
y2[0]=1
y2_diff = [steps]
y2_diff[0] = 0
k1=0
k2=0

#implizites Euler-Verfahren
y3=[steps]
y3[0]=1
y3_diff = [steps]
y3_diff[0] = 0

#analytische Losung
y4=[steps]
y4[0]=1


for i in range(0, int(steps)-1):
	#explizites Euler-Verfahren
	y1.append(y1[i]+h*y_dot(x[i],y1[i]));
   
	#RK2-Verfahren
	k1 = h*y_dot(x[i],y2[i])
	k2 = h*y_dot(x[i]+(h/2),y2[i]+(k1/2))
	y2.append(y2[i]+k2)

	#implizites Euler-Verfahren _TODO
	#implizites Euler-Verfahren
	y3.append((y3[i] + 10*h + 5000*(x[i] + h)*h) / (1 + 500*h))

	#analytische Losung
	y4.append(y(x[i]))
   
	x.append(x[i]+h)


for i in range(0, int(steps)-1):
	y1_diff.append(abs(y4[i]-y1[i]))
	y2_diff.append(abs(y4[i]-y2[i]))
	y3_diff.append(abs(y4[i]-y3[i]))

    
plt.plot(x,y1, 'r') 
plt.plot(x,y2, 'b')  
plt.plot(x,y3, 'g') 
plt.plot(x,y4, 'm') 
plt.title('Alle Verfahren')
plt.savefig("combine.png")
#plt.legend()
plt.show()    
    
plt.plot(x,y1_diff, 'r') 
plt.plot(x,y2_diff, 'b')  
plt.plot(x,y3_diff, 'g') 
plt.title('Alle Verfahren')
plt.savefig("difference.png")
plt.show()
#plt.close()
```

```python
import matplotlib.pyplot as plt
import math

def y1_dot(y1,y2):
   return (6*y1 - 6*y1*y2*y2 - y2)


def y2_dot(y1):
   return (y1)


h = 0.02
x_End = 31

steps = x_End/h

x = [steps]
x[0]=0

#explizites Euler-Verfahren
y1_e = [steps]
y2_e = [steps]
y1_e[0]=0
y2_e[0]=1

#RK2-Verfahren
y1_rk=[steps]
y2_rk=[steps]
y1_rk[0]=0
y2_rk[0]=1
k1=0
k2=0
l1=0
l2=0

for i in range(0, int(steps)-1):
	#explizites Euler-Verfahren
	y1_e.append(y1_e[i] + h*y1_dot(y1_e[i],y2_e[i]));
	y2_e.append(y2_e[i] + h*y2_dot(y1_e[i]));    
    
   
	#RK2-Verfahren
	k1 = h*y1_dot(y1_rk[i],y2_rk[i])
	l1 = h*y1_rk[i]
    
	k2 = h*y1_dot((y1_rk[i]+(k1/2)),(y2_rk[i]+(l1/2)))          
	l2 = h*y2_dot((y1_rk[i]+(k1/2)))
                  
	y1_rk.append(y1_rk[i]+k2)
	y2_rk.append(y2_rk[i]+l2)
                  
	x.append(x[i]+h)
    
plt.plot(x,y2_e, 'r') 
plt.plot(x,y2_rk, 'b')  
plt.title('DGL 2.Ordnung')
plt.savefig("dgl_2ord.png")
plt.show()    
#plt.close()
```

```python
import matplotlib.pyplot as plt
import math

def x_dot(x,y):
   return (-10*(x-y))

def y_dot(x,y,z):
   return ((40-z)*x - y)

def y2_dot(x,y,z):
   return ((40.000000001-z)*x - y)

def z_dot(x,y,z):
   return (x*y - 2.67*z)


h = 0.002
x_End = 120

steps = x_End/h

x_time = [steps]
x_time[0]=0


#RK2-Verfahren
x=[steps]
x[0]=0.01
k1_x=0
k2_x=0

y=[steps]
y[0]=0.01
k1_y=0
k2_y=0

z=[steps]
z[0]=0
k1_z=0
k2_z=0

x2=[steps]
x2[0]=0.01
k1_x2=0
k2_x2=0

y2=[steps]
y2[0]=0.01
k1_y2=0
k2_y2=0

z2=[steps]
z2[0]=0
k1_z2=0
k2_z2=0


for i in range(0, int(steps)-1):
	k1_x = h*x_dot(x[i],y[i]) 
	k2_x = h*x_dot((x[i]+(k1_x/2)),y[i])
	x.append(x[i]+k2_x)
                    
	k1_y = h*y_dot(x[i],y[i],z[i]) 
	k2_y = h*y_dot(x[i],(y[i]+(k1_y/2)),z[i])
	y.append(y[i]+k2_y)
    
	k1_z = h*z_dot(x[i],y[i],z[i]) 
	k2_z = h*z_dot(x[i],y[i],(z[i]+(k1_z/2)))
	z.append(z[i]+k2_z)
  
	k1_x2 = h*x_dot(x2[i],y2[i]) 
	k2_x2 = h*x_dot((x2[i]+(k1_x2/2)),y2[i])
	x2.append(x2[i]+k2_x2)
    
	k1_y2 = h*y2_dot(x2[i],y2[i],z2[i]) 
	k2_y2 = h*y2_dot(x2[i],(y2[i]+(k1_y2/2)),z2[i])
	y2.append(y2[i]+k2_y2)
    
	k1_z2 = h*z_dot(x2[i],y2[i],z2[i]) 
	k2_z2 = h*z_dot(x2[i],y2[i],(z2[i]+(k1_z2/2)))
	z2.append(z2[i]+k2_z2)
                      
	x_time.append(x_time[i]+h)
    
plt.plot(x_time,x, 'r') 
plt.title('Lorenz-Attraktor x(t) 40')
plt.savefig("lorenz_xt40.png")
plt.show() 

plt.plot(x_time,x2, 'r') 
plt.title('Lorenz-Attraktor x(t) 40.00..1')
plt.savefig("lorenz_zx401.png")
plt.show()

plt.plot(x,z, 'r') 
plt.title('Lorenz-Attraktor z(x)')
plt.savefig("lorenz_zx.png")
plt.show()
#plt.close()
```
